```python
import pandas as pd
from math import radians, cos, sin, asin, sqrt
import folium
import xml.etree.ElementTree as ET
gps_tree = ET.parse('session/Kismet-20191011-18-31-55-1.gpsxml')
gps_root = gps_tree.getroot()
```

```python
coordinates = []
for child in gps_root:
    if (child.tag == "gps-point"):
        coordinates.append([float(child.attrib['lat']),float(child.attrib['lon'])])
coordinates = list(set(map(tuple,coordinates)))
print(len(coordinates))
#print(coordinates)
```

```python
def haversine(lon1, lat1, lon2, lat2): # haversine formula for point distance
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    m = 6371* c/ 1.069
    return m
```
